```python
#NYT cases data
us_counties_cases = pd.read_csv("us-counties.csv")
#us_counties_cases.set_index('date',inplace=True)
us_counties_cases.reset_index(drop=True, inplace=True)
us_counties_cases
```

```python
# cases only for 1 day to match common counties NYT
# start= "2020-12-29"
# end = "2020-12-30"

start= "2020-08-01"
end = "2020-12-31"

us_counties_cases['date'] = pd.to_datetime(us_counties_cases['date'])
us_counties_cases.reset_index(drop=True, inplace=True)

#greater than the start date and smaller than the end date
mask = (us_counties_cases['date'] > start) & (us_counties_cases['date'] <= end)

us_counties_cases_dec = us_counties_cases.loc[mask]
us_counties_cases_dec.reset_index(drop=True, inplace=True)
us_counties_cases_dec
```

```python
# us_counties_cases_new = us_counties_cases_dec.iloc[0:3244]
# #us_counties_cases_new

us_counties_cases = us_counties_cases_dec[['date', 'county', 'cases']].copy()
#us_counties_cases

us_counties_cases.reset_index(drop=True, inplace=True)
us_counties_cases = us_counties_cases
us_counties_cases
```

```python
county_list = []
for counties in us_counties_cases_dec["county"]:
    county_list.append(counties)
#print (len(county_list))

df_time_series = pd.DataFrame(columns = county_list)
df_time_series
```

```python
# Create the cases/predictor(Z) variable 

county_num = 3245
row = 0
for index in range(us_counties_cases.shape[0]):    
    county_found = us_counties_cases["county"].iloc[index]
    if index <= county_num:
        df_time_series.loc[row,county_found] = us_counties_cases.loc[index,"cases"]
        
    else:
        county_num = (county_num * 1) + 3245
        row = row + 1
    
df_time_series
        
    
```

```python
# Replace the missing values with the previosu day
for column in df_time_series.columns :
    for index, row in df_time_series.iterrows():
        if index >0 :
            if (pd.isnull(df_time_series.loc[index,column])) is True:
                #print ("Previous index:", index - 1)
                #print ("Index previous value is:",df_time_series.loc[index -1,column])
                df_time_series.loc[index,column] = df_time_series.loc[index -1,column]
```
